```python
cor = data.corr().round(2)

# Mask for the upper traingle of chart
mask_up = np.zeros_like(cor, dtype=np.bool)
mask_up[np.triu_indices_from(mask_up)] = True

# Setting figure size for graph
fig, axis = plt.subplots(figsize=(20, 20))

# Defining custom color for map
color_map = sns.diverging_palette(221, 9, as_cmap=True)

# heatmap using seaborn
sns.heatmap(cor, mask=mask_up, cmap=color_map, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.7, cbar_kws={"shrink": .7}, annot=True)

plt.tight_layout()
```

```python
# dropping all "worst" columns
colms = ['radius_worst', 
        'texture_worst', 
        'perimeter_worst', 
        'area_worst', 
        'smoothness_worst', 
        'compactness_worst', 
        'concavity_worst',
        'concave points_worst', 
        'symmetry_worst', 
        'fractal_dimension_worst']
data = data.drop(colms, axis=1)

# dropping all columns about "perimeter" and "area"
colms = ['perimeter_mean',
        'perimeter_se', 
        'area_mean', 
        'area_se']
data = data.drop(colms, axis=1)

# dropping all columns about "concavity" and "concave points"
colms = ['concavity_mean',
        'concavity_se', 
        'concave points_mean', 
        'concave points_se']
data = data.drop(colms, axis=1)

# checking all remaining columns
data.columns
```

```python
# heatmap with the new correlation matrix
cor = data.corr().round(2)
mask2 = np.zeros_like(cor, dtype=np.bool)
mask2[np.triu_indices_from(mask2)] = True

fig, axis = plt.subplots(figsize=(20, 20))
sns.heatmap(cor, mask=mask2, cmap=color_map, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.7, cbar_kws={"shrink": .7}, annot=True)
plt.tight_layout()
```
